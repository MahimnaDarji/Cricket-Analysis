```python
dot=data[data.runs_off_bat == 0]
```

```python
d1=dot.groupby('striker')['ball'].count().sort_values(ascending=False).head(10)
d1
```

```python
d2=data.groupby('striker')['ball'].count().sort_values(ascending=False).head(10)
d2
```

```python
d1=pd.DataFrame(dot.groupby('striker')['ball'].count().sort_values(ascending=False)).reset_index()
d2=pd.DataFrame(data.groupby('striker')['ball'].count().sort_values(ascending=False)).reset_index()
```

```python
d3=d1.merge(d2,on='striker')
d3.head(10)
```

```python
d3['dot']=(100*d3['ball_x']/d3['ball_y'])
d3=d3[(d3.ball_y >= 2000)]
d3
```

```python
boun=data[(data.runs_off_bat == 4) | (data.runs_off_bat == 6)]
boun
```

```python
b1=boun.groupby('striker')['runs_off_bat'].count().sort_values(ascending=False).head(10)
b1
```

```python
d2=data.groupby('striker')['ball'].count().sort_values(ascending=False).head(10)
d2
```

```python
b1=pd.DataFrame(boun.groupby('striker')['runs_off_bat'].count().sort_values(ascending=False)).reset_index()
d2=pd.DataFrame(data.groupby('striker')['ball'].count().sort_values(ascending=False)).reset_index()
```

```python
b3=b1.merge(d2,on='striker')
b3
```

```python
b3['boun']=(100*b3['runs_off_bat']/b3['ball'])
b3=b3[b3.ball > 2000]
b3
```

```python
combineddata=b3.merge(d3,on='striker')
combineddata
```

```python
combineddata=combineddata.drop(['runs_off_bat','ball','ball_x','ball_y'],axis=1)
```

```python
combineddata
```

```python
x = combineddata['striker']
y = combineddata['boun']
  
# first plot with X and Y data
plt.plot(x, y)
  
x1 = combineddata['striker']
y1 = combineddata['dot']
  
# second plot with x1 and y1 data
plt.plot(x1, y1, '-.')
  
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')
plt.show()
```
