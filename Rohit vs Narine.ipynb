```python
r1=pd.DataFrame(rohit.groupby('bowler')['runs_off_bat'].sum().sort_values(ascending=False)).reset_index()
```

```python
r4=r1.merge(r2,on='bowler',how='left')
r4
```

```python
r5=r4.merge(r3,on='bowler',how='left')
r5['StrikeRate']=100*r5['runs_off_bat']/r5['ball']
r5=r5[(r5.player_dismissed >= 3) & (r5.ball >= 40)]
r5['Average']=r5['runs_off_bat']/r5['player_dismissed']
r5['StrPAvg']=r5['StrikeRate'] + r5['Average']
r5
```

```python
plt.figure(figsize=(16,8))
size=[r5.player_dismissed * 50]
plt.scatter(r5.Average,r5.StrikeRate,s=size,alpha=1)
plt.title('Rohit StrPAvg Vs Bowlers he struggles most')
plt.axvline(16,ls='--',color='red')
plt.axhline(100,ls='--',color='red')
plt.show()
```
