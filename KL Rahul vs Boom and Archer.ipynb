```python
b1=pd.DataFrame(bum.groupby('striker')['runs_off_bat'].sum().sort_values(ascending = False)).reset_index()
b2=pd.DataFrame(bum.groupby('striker')['player_dismissed'].count().sort_values(ascending = False)).reset_index()
b3=pd.DataFrame(bum.groupby('striker')['ball'].count().sort_values(ascending = False)).reset_index()
```

```python
b4=b1.merge(b3,on='striker')
b4
```

```python
b5=b4.merge(b2,on='striker')
b5.head(10)
```

```python
b4.reset_index(inplace=True,drop=True)
b5.reset_index(inplace=True,drop=True)
```

```python
b5['StrikeRate']=100*(b5['runs_off_bat']/b5['ball'])
b5=b5[(b5.runs_off_bat > 30)]
b5
```

```python
plt.figure(figsize=(16,8))
plt.scatter(b5.runs_off_bat,b5.StrikeRate)

for i in range(len(b5)):
    plt.text(b5['runs_off_bat'][i],b5['StrikeRate'][i],b5['striker'][i])
plt.xlabel('Runs Scored')
plt.ylabel('StrikeRate')
plt.axvline(70,ls='--')
plt.axhline(130,ls='--')
plt.show()
```
