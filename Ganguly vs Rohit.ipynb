```python
data=data[data['Innings Runs Scored Num'] != '-']
```

```python
data=data.dropna(subset=['Innings Runs Scored Num'])
```

```python
data['Innings Runs Scored Num'].unique()
```

```python
data['Innings Date']=pd.to_datetime(data['Innings Date'])
```

```python
data['Year']=data['Innings Date'].dt.year
```

```python
data.head(2)
```

```python
data['Innings Runs Scored Num'] = data['Innings Runs Scored Num'].astype('int')
```

```python
data['Innings Balls Faced'] = data['Innings Balls Faced'].astype('int')
```

```python
data['Innings Not Out Flag'] = data['Innings Not Out Flag'].astype('float')
```

```python
ganguly_data=data[(data.Year >= 1997) & (data.Year <= 2007)]
```

```python
rohit_data=data[(data.Year >= 2009) & (data.Year <= 2019)]
```

```python
rohit_data.head(2)
```

```python
ganguly_data.tail(20)
```

```python
data.dtypes
```

```python
#For Ganguly
sg=ganguly_data[ganguly_data['Innings Player'] == 'SC Ganguly']
```

```python
sg.head(5)
```

```python
#For Rohit Sharma
rg=rohit_data[rohit_data['Innings Player'] == 'RG Sharma']
```

```python
rg.tail(5)
```

```python
#Runs Scored by Ganguly between 1997 to 2007
print("Total Runs of Ganguly",sum(sg['Innings Runs Scored Num']))
print("Total Inning of Ganguly",sum(sg['Innings Batted Flag']))
```

```python
#Runs Scored by Rohit Sharma between 2009 to 2019
print("Total Runs of Rohit",sum(rg['Innings Runs Scored Num']))
print("Total Innings of Rohit",sum(rg['Innings Batted Flag']))
```

```python
#Runs Per Innings of Both Ganguly and Rohit
print("RPI Of Ganguly",(sum(sg['Innings Runs Scored Num'])/sum(sg['Innings Batted Flag'])))
print("RPI Of Rohit",(sum(rg['Innings Runs Scored Num'])/sum(rg['Innings Batted Flag'])))
```

```python
#Strike Rate of Both Ganguly and Rohit
print("Strike Rate of Ganguly",(100*(sum(sg['Innings Runs Scored Num']))/sum(sg['Innings Balls Faced'])))
print("Strike Rate of Rohit",(100*(sum(rg['Innings Runs Scored Num']))/sum(rg['Innings Balls Faced'])))
```

```python
#No. Of Hundreds in their Peak and Innings Per Hundred
print("Hundreds of Ganguly",sum(sg["100's"]))
print("Hundreds of Rohit",sum(rg["100's"]))
print("Innings Per Hundred Of Ganguly",(sum(sg['Innings Batted Flag']))/sum(sg["100's"]))
print("Innings Per Hundred Of Rohit",(sum(rg['Innings Batted Flag']))/sum(rg["100's"]))
```

```python
#No. Of Fifties in their Peak and Innings Per Hundred
print("Fifties of Ganguly",sum(sg["50's"]))
print("Fifties of Rohit",sum(rg["50's"]))
print("Innings Per Fifties Of Ganguly",(sum(sg['Innings Batted Flag']))/sum(sg["50's"]))
print("Innings Per Fifties Of Rohit",(sum(rg['Innings Batted Flag']))/sum(rg["50's"]))
```

```python
#Total Contribution of Ganguly in Teams Score
print("Total Team Runs",sum(ganguly_data[ganguly_data.Country == 'India']['Innings Runs Scored Num']))
print("Ganguly's Total Runs",sum(sg['Innings Runs Scored Num']))
print("Contribution Of Ganguly in Team's Run",(100*sum(sg['Innings Runs Scored Num']))/(sum(ganguly_data[ganguly_data.Country == 'India']['Innings Runs Scored Num'])),"%")
```

```python
#Total Contribution of Rohit in Teams Score
print("Total Team Runs",sum(rohit_data[rohit_data.Country == 'India']['Innings Runs Scored Num']))
print("Rohit's Total Runs",sum(rg['Innings Runs Scored Num']))
print("Contribution Of Rohit in Team's Run",(100*sum(rg['Innings Runs Scored Num']))/(sum(rohit_data[rohit_data.Country == 'India']['Innings Runs Scored Num'])),"%")
```

```python
# Average Number of Balls Faced per Innings
print("Average Number of Balls Faced by Ganguly",(sum(sg['Innings Balls Faced']))/sum(sg['Innings Batted Flag']))
print("Average Number of Balls Faced by Rohit",(sum(rg['Innings Balls Faced']))/sum(rg['Innings Batted Flag']))
```
