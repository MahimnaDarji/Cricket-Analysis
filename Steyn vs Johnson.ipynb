```python
johnson_data=data[(data.Year >= 2005) & (data.Year <= 2015)]
```

```python
steyn_data=data[(data.Year >= 2005) & (data.Year <= 2019)]
```

```python
steyn_data.head(5)
```

```python
mj=johnson_data[johnson_data['Innings Player'] == 'MG Johnson']
```

```python
mj.head(5)
```

```python
ds=steyn_data[steyn_data['Innings Player'] == 'DW Steyn']
```

```python
ds.head(5)
```

```python
#Total Matches Played
print("Total Matches Played By Johnson",print("Total Matches Played By Johnson",sum(mj['Innings Bowled Flag'])))
print("Total Matches Played By Steyn",sum(ds['Innings Bowled Flag']))
```

```python
#Total Wickets Taken By Both
print("Total Wickets Of Johnson",sum(mj['Innings Wickets Taken']))
print("Total Wickets Of Steyn",sum(ds['Innings Wickets Taken']))
```

```python
#Wickets Per Innings
print("WPI Of Johnson",(sum(mj['Innings Wickets Taken'])/sum(mj['Innings Bowled Flag'])))
print("WPI Of Steyn",(sum(ds['Innings Wickets Taken'])/sum(ds['Innings Bowled Flag'])))
```

```python
#Career Economy Rate
print("Economy Rate Of Johnson",(sum(mj['Innings Runs Conceded'])/sum(mj['Innings Overs Bowled'])))
print("Economy Rate Of Steyn",(sum(ds['Innings Runs Conceded'])/sum(ds['Innings Overs Bowled'])))
```

```python
#Maiden Overs
print("Total Maiden Overs Bowled By Johnson",sum(mj['Innings Maidens Bowled']))
print("Total Maiden Overs Bowled By Steyn",sum(ds['Innings Maidens Bowled']))
print("Number Of Maidens per Over of Johnson",sum(mj['Innings Overs Bowled'])/sum(mj['Innings Maidens Bowled']))
print("Number Of Maidens per Over of Steyn",sum(ds['Innings Overs Bowled'])/sum(ds['Innings Maidens Bowled']))
```

```python
#Bowling Average
print("Bowling Average of Johnson",(sum(mj['Innings Runs Conceded'])/sum(mj['Innings Wickets Taken'])))
print("Bowling Average of Steyn",(sum(ds['Innings Runs Conceded'])/sum(ds['Innings Wickets Taken'])))
```

```python
#Contribution in their Teams Wickets
print("Total Wickets Taken By Australia During Johnson's Career",(sum(johnson_data[johnson_data.Country == 'Australia']['Innings Wickets Taken'])))
print("Total Wickets Taken By Johnson in his Career",sum(mj['Innings Wickets Taken']))
print("Contribution Of Johnson in Team's Wickets",(100*sum(mj['Innings Wickets Taken']))/(sum(johnson_data[johnson_data.Country == 'Australia']['Innings Wickets Taken'])),"%")
```

```python
print("Total Wickets Taken By Soth Africa During Steyn's Career",(sum(steyn_data[steyn_data.Country == 'South Africa']['Innings Wickets Taken'])))
print("Total Wickets Taken By Steyn in his Career",sum(ds['Innings Wickets Taken']))
print("Contribution Of Steyn in Team's Wickets",(100*sum(ds['Innings Wickets Taken']))/(sum(johnson_data[johnson_data.Country == 'South Africa']['Innings Wickets Taken'])),"%")
```
