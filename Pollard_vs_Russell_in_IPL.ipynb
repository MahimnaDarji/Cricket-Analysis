```python
print("Total Runs=",sum(pol['runs_off_bat']))
print("Balls Faced=",len(pol))
print("Strike Rate=",100*(sum(pol['runs_off_bat'])/len(pol)))
print("Times Dismissed=",sum(pol.player_dismissed == 'KA Pollard'))
print("Average=",(sum(pol['runs_off_bat'])/sum(pol.player_dismissed == 'KA Pollard')))
```

#Percentage Of Team's Runs Scored During their Playing Time

```python
mi=data[data.batting_team == 'Mumbai Indians']
mi=mi[(mi.innings == 1) | (mi.innings == 2)]
mi=mi[(mi.season >= 2010) & (mi.season <= 2022)]
```

```python
mi.tail(5)
```

```python
kkr=data[data.batting_team == 'Kolkata Knight Riders']
kkr=kkr[(kkr.innings == 1) | (kkr.innings == 2)]
kkr=kkr[(kkr.season >= 2014) & (kkr.season <= 2022)]
```

```python
kkr.head(5)
```

```python
print("Total Runs Scored by Mumbai Indians After Pollard's Debut=",sum(mi['runs_off_bat']))
print("Runs Scored by Pollard=",sum(pol['runs_off_bat']))
print("Scored Percentage of Team's Runs=",100*(sum(pol['runs_off_bat'])/sum(mi['runs_off_bat'])),"%")
```

```python
print("Total Runs Scored by KKR After Russell's Debut=",sum(kkr['runs_off_bat']))
print("Runs Scored by Russell=",sum(russ['runs_off_bat']))
print("Scored Percentage of Team's Runs=",100*(sum(russ['runs_off_bat'])/sum(kkr['runs_off_bat'])),"%")
```

####Observation:- Pollard has contributed more towards teams runs then Russell
